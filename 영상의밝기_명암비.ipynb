```python
# 히스토그램 스트레칭
import numpy as np
import cv2 as cv

src = cv.imread('c:/ex/cv/hawkes.bmp', cv.IMREAD_GRAYSCALE)
dst = ((src-np.min(src))*255. / (np.max(src) - np.min(src))).astype(np.uint8)

import matplotlib.pyplot as plt

s = cv.calcHist([src],[0],None,[256],[0,256])
d = cv.calcHist([dst],[0],None,[256],[0,256])
plt.plot(s,'k',d,'k')
plt.show()

cv.imshow('src',src)
cv.imshow('dst',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```
