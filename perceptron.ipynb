```python
import math
import numpy as np

# 파이썬에서 자연상수
print(f"math.e = {math.e}")
print(f"exp(1) = {math.exp(1)}")

# 극한 정의 확인
n_values = [10, 100, 1000, 10000, 100000]
for n in n_values:
    approx = (1 + 1/n) ** n
    print(f"n={n:6d}: (1+1/n)^n = {approx:.10f}")
```

```python
# 오일러-마스케로니 상수 근사
def euler_mascheroni_approx(n):
    harmonic_sum = sum(1/k for k in range(1, n+1))
    return harmonic_sum - math.log(n)

# 계산 예시
for n in [100, 1000, 10000, 100000]:
    gamma_approx = euler_mascheroni_approx(n)
    print(f"n={n:6d}: γ ≈ {gamma_approx:.10f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification, load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from collections import Counter
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = ['Malgun Gothic', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

class KNNFromScratch:
    """k-NN 알고리즘 직접 구현"""
    
    def __init__(self, k=3):
        self.k = k
        
    def fit(self, X, y):
        """훈련 데이터 저장 (실제로는 저장만 함)"""
        self.X_train = X
        self.y_train = y
        
    def euclidean_distance(self, point1, point2):
        """유클리드 거리 계산"""
        return np.sqrt(np.sum((point1 - point2) ** 2))
    
    def predict(self, X):
        """예측 수행"""
        predictions = []
        
        for test_point in X:
            # 모든 훈련 데이터와의 거리 계산
            distances = []
            for i, train_point in enumerate(self.X_train):
                dist = self.euclidean_distance(test_point, train_point)
                distances.append((dist, self.y_train[i]))
            
            # 거리순으로 정렬하여 가장 가까운 k개 선택
            distances.sort(key=lambda x: x[0])
            k_nearest = distances[:self.k]
            
            # k개 이웃의 라벨 중 다수결로 예측
            k_nearest_labels = [label for _, label in k_nearest]
            prediction = Counter(k_nearest_labels).most_common(1)[0][0]
            predictions.append(prediction)
            
        return np.array(predictions)

def visualize_knn_2d():
    """2D에서 k-NN 시각화"""
    # 간단한 2D 데이터 생성
    np.random.seed(42)
    
    # 클래스 A (빨간색)
    class_A = np.random.normal([2, 2], 0.8, (20, 2))
    # 클래스 B (파란색)  
    class_B = np.random.normal([6, 6], 0.8, (20, 2))
    
    X = np.vstack([class_A, class_B])
    y = np.array([0]*20 + [1]*20)
    
    # 새로운 테스트 포인트
    test_point = np.array([4, 4])
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for idx, k in enumerate([1, 3, 7]):
        ax = axes[idx]
        
        # k-NN 모델 훈련
        knn = KNNFromScratch(k=k)
        knn.fit(X, y)
        
        # 거리 계산
        distances = []
        for i, point in enumerate(X):
            dist = knn.euclidean_distance(test_point, point)
            distances.append((dist, i, y[i]))
        
        distances.sort()
        k_nearest_indices = [idx for _, idx, _ in distances[:k]]
        
        # 시각화
        colors = ['red', 'blue']
        for class_idx in [0, 1]:
            mask = y == class_idx
            ax.scatter(X[mask, 0], X[mask, 1], c=colors[class_idx], 
                      alpha=0.6, s=50, label=f'클래스 {class_idx}')
        
        # k개 최근접 이웃 강조
        for i in k_nearest_indices:
            ax.scatter(X[i, 0], X[i, 1], c='black', s=100, marker='x')
            # 거리선 그리기
            ax.plot([test_point[0], X[i, 0]], [test_point[1], X[i, 1]], 
                   'k--', alpha=0.5)
        
        # 테스트 포인트
        ax.scatter(test_point[0], test_point[1], c='green', s=200, 
                  marker='*', label='테스트 포인트')
        
        # 예측 결과
        prediction = knn.predict([test_point])[0]
        ax.set_title(f'k={k}, 예측: 클래스 {prediction}')
        ax.legend()
        ax.grid(True, alpha=0.3)
        ax.set_xlabel('특성 1')
        ax.set_ylabel('특성 2')
    
    plt.tight_layout()
    plt.show()

def compare_distance_metrics():
    """다양한 거리 측정 방법 비교"""
    
    # 2D 포인트들
    point1 = np.array([1, 1])
    point2 = np.array([4, 5])
    
    # 거리 계산 함수들
    def euclidean_distance(p1, p2):
        return np.sqrt(np.sum((p1 - p2) ** 2))
    
    def manhattan_distance(p1, p2):
        return np.sum(np.abs(p1 - p2))
    
    def chebyshev_distance(p1, p2):
        return np.max(np.abs(p1 - p2))
    
    # 거리 계산
    distances = {
        '유클리드': euclidean_distance(point1, point2),
        '맨하탄': manhattan_distance(point1, point2),
        '체비셰프': chebyshev_distance(point1, point2)
    }
    
    print("=== 거리 측정 방법 비교 ===")
    print(f"포인트 1: {point1}")
    print(f"포인트 2: {point2}")
    for name, dist in distances.items():
        print(f"{name} 거리: {dist:.2f}")
    
    # 시각화
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # 점들 표시
    ax.scatter(*point1, c='red', s=100, label='포인트 1')
    ax.scatter(*point2, c='blue', s=100, label='포인트 2')
    
    # 거리선들 표시
    ax.plot([point1[0], point2[0]], [point1[1], point2[1]], 
           'g-', linewidth=2, label='유클리드 거리')
    
    # 맨하탄 거리 (직각 경로)
    ax.plot([point1[0], point2[0], point2[0]], 
           [point1[1], point1[1], point2[1]], 
           'r--', linewidth=2, label='맨하탄 거리')
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('거리 측정 방법들')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()

def k_value_analysis():
    """k 값에 따른 성능 분석"""
    
    # 아이리스 데이터셋 사용
    iris = load_iris()
    X, y = iris.data, iris.target
    
    # 훈련/테스트 분할
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42)
    
    # 다양한 k 값에 대해 성능 측정
    k_range = range(1, 21)
    train_accuracies = []
    test_accuracies = []
    
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        
        train_pred = knn.predict(X_train)
        test_pred = knn.predict(X_test)
        
        train_acc = accuracy_score(y_train, train_pred)
        test_acc = accuracy_score(y_test, test_pred)
        
        train_accuracies.append(train_acc)
        test_accuracies.append(test_acc)
    
    # 시각화
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, train_accuracies, 'bo-', label='훈련 정확도')
    plt.plot(k_range, test_accuracies, 'ro-', label='테스트 정확도')
    plt.xlabel('k 값')
    plt.ylabel('정확도')
    plt.title('k 값에 따른 성능 변화')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 최적 k 값 표시
    best_k = k_range[np.argmax(test_accuracies)]
    best_acc = max(test_accuracies)
    plt.axvline(x=best_k, color='green', linestyle='--', 
               label=f'최적 k={best_k} (정확도={best_acc:.3f})')
    plt.legend()
    plt.show()
    
    print(f"최적 k 값: {best_k}")
    print(f"최고 테스트 정확도: {best_acc:.3f}")

def decision_boundary_visualization():
    """결정 경계 시각화"""
    
    # 2차원 데이터 생성
    X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, 
                              n_informative=2, n_clusters_per_class=1, 
                              random_state=42)
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for idx, k in enumerate([1, 5, 15]):
        ax = axes[idx]
        
        # k-NN 모델 훈련
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X, y)
        
        # 결정 경계 그리기
        h = 0.02
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                            np.arange(y_min, y_max, h))
        
        Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        # 결정 경계 시각화
        ax.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.RdYlBu)
        
        # 데이터 포인트 표시
        scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='black')
        
        ax.set_title(f'k={k}의 결정 경계')
        ax.set_xlabel('특성 1')
        ax.set_ylabel('특성 2')
    
    plt.tight_layout()
    plt.show()

def bias_variance_analysis():
    """편향-분산 트레이드오프 분석"""
    
    print("=== k-NN의 편향-분산 트레이드오프 ===")
    print("k가 작을 때 (k=1):")
    print("  - 편향(Bias): 낮음 - 데이터에 매우 민감하게 반응")
    print("  - 분산(Variance): 높음 - 노이즈에 과적합 경향")
    print("  - 결정 경계: 복잡하고 불규칙함")
    print()
    print("k가 클 때 (k=전체):")
    print("  - 편향(Bias): 높음 - 지나치게 단순화")
    print("  - 분산(Variance): 낮음 - 안정적이지만 과소적합")
    print("  - 결정 경계: 매우 단순함")
    print()
    print("최적 k 선택:")
    print("  - 교차 검증을 통한 최적값 찾기")
    print("  - 일반적으로 k = sqrt(n) 정도를 시작점으로 사용")

def knn_advantages_disadvantages():
    """k-NN 장단점 분석"""
    
    advantages = [
        "단순하고 이해하기 쉬움",
        "훈련 시간이 매우 빠름 (게으른 학습)",
        "비선형 결정 경계 자연스럽게 처리",
        "새로운 데이터에 즉시 적응 가능",
        "분류와 회귀 모두 가능",
        "데이터 분포에 대한 가정 불필요"
    ]
    
    disadvantages = [
        "예측 시간이 느림 (모든 거리 계산 필요)",
        "차원의 저주에 민감함",
        "불균형 데이터에 취약함",
        "노이즈와 이상치에 민감함",
        "특성 스케일링이 중요함",
        "메모리 사용량이 많음"
    ]
    
    print("=== k-NN 알고리즘 분석 ===")
    print("\n🟢 장점:")
    for i, adv in enumerate(advantages, 1):
        print(f"  {i}. {adv}")
    
    print("\n🔴 단점:")
    for i, dis in enumerate(disadvantages, 1):
        print(f"  {i}. {dis}")

def practical_example():
    """실제 적용 예시"""
    
    print("=== k-NN 실제 활용 예시 ===")
    
    # 아이리스 데이터로 실제 예시
    iris = load_iris()
    X, y = iris.data, iris.target
    
    # 데이터 분할
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42)
    
    # 여러 k 값으로 모델 비교
    k_values = [1, 3, 5, 7, 9]
    
    print("\nk 값별 성능 비교:")
    print("-" * 50)
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        
        train_acc = knn.score(X_train, y_train)
        test_acc = knn.score(X_test, y_test)
        
        print(f"k={k:2d} | 훈련 정확도: {train_acc:.3f} | 테스트 정확도: {test_acc:.3f}")
    
    # 최적 모델로 상세 분석
    best_knn = KNeighborsClassifier(n_neighbors=5)
    best_knn.fit(X_train, y_train)
    y_pred = best_knn.predict(X_test)
    
    print(f"\n=== 최적 모델 (k=5) 상세 결과 ===")
    print(f"테스트 정확도: {accuracy_score(y_test, y_pred):.3f}")
    print("\n분류 보고서:")
    print(classification_report(y_test, y_pred, target_names=iris.target_names))

# 모든 함수 실행
if __name__ == "__main__":
    print("k-NN 알고리즘 완전 분석\n")
    
    # 1. 기본 개념 시각화
    print("1. k-NN 기본 개념 시각화")
    visualize_knn_2d()
    
    # 2. 거리 측정 방법 비교
    print("\n2. 거리 측정 방법 비교")
    compare_distance_metrics()
    
    # 3. k 값 분석
    print("\n3. k 값에 따른 성능 분석")
    k_value_analysis()
    
    # 4. 결정 경계 시각화
    print("\n4. 결정 경계 시각화")
    decision_boundary_visualization()
    
    # 5. 편향-분산 분석
    print("\n5. 편향-분산 트레이드오프")
    bias_variance_analysis()
    
    # 6. 장단점 분석
    print("\n6. 장단점 분석")
    knn_advantages_disadvantages()
    
    # 7. 실제 예시
    print("\n7. 실제 활용 예시")
    practical_example()
```
